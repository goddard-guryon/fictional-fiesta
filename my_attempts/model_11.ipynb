# Density-Based Clustering

To cluster weather stations in Canada based on location and average temperature

Import necessary modules (make sure you have installed `GEOS` from `github.com/libgeos/geos` and `Basemap` via `apt-get install python3-mpltoolkits.basemap` or from `github.com/matplotlib/basemap` before running this):

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
import pickle

from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

%matplotlib inline
```

Get the dataset and see a sample:

```python
df = pd.read_csv("../datasets/weather-stations20140101-20141231.csv")
df.head()
```

And description of the dataset:

```python
df.describe()
```

Process the dataset to remove null values of temperature:

```python
df = df[pd.notnull(df["Tm"])].reset_index(drop=True)
df.head()
```

For visualizing the map, I have created a function below which would work on both uncluestered and clustered datasets:

```python
def draw_map(df, a, b, c, d, clustered=False, n_clusters=0, model=None):
    from pylab import rcParams
    rcParams["figure.figsize"] = (14, 10)
    df = df.copy()

    lat_up, lat_dn, lon_up, lon_dn = a, b, c, d  # 65, 40, -50, -140
    df = df[(df["Lat"] > lat_dn) &
            (df["Lat"] < lat_up) &
            (df["Long"] > lon_dn) &
            (df["Long"] < lon_up)]
    data = Basemap(projection="merc",
                  resolution='l',
                  area_thresh=1000.0,
                  llcrnrlat=lat_dn,
                  urcrnrlat=lat_up,
                  llcrnrlon=lon_dn,
                  urcrnrlon=lon_up)
    data.drawcoastlines()
    data.drawcountries()
    data.fillcontinents(color="white", alpha=0.3)
    data.shadedrelief()
    
    xs, ys = data(np.asarray(df["Long"]), np.asarray(df["Lat"]))
    df["xm"], df["ym"] = xs.tolist(), ys.tolist()
    
    if clustered:
        colors = plt.get_cmap("jet")(np.linspace(0, 1, n_clusters))
        for clust_num in set(model.labels_):
            c = (([0.4, 0.4, 0.4]) if clust_num==-1 else colors[np.int(clust_num)])
            cluster = df[df["labels"]==clust_num]
            data.scatter(cluster["xm"],
                        cluster["ym"],
                        color=c,
                        marker='o',
                        s=20,
                        alpha=0.85,
                        linewidth=0)
            if clust_num != -1:
                center_x = np.mean(cluster["xm"])
                center_y = np.mean(cluster["ym"])
                plt.text(center_x,
                        center_y,
                        str(clust_num),
                        fontsize=25,
                        color="red")
                print(f"Cluster {clust_num}, Average temperature: {np.mean(cluster['Tm'])}")
    else:
        for index, row in df.iterrows():
            data.plot(df["xm"],
                     df["ym"],
                     markerfacecolor=([1, 0, 0]),
                     marker='o',
                     markersize=3,
                     alpha=0.75,
                     linewidth=0)
    plt.show()
    return df
```

Let's look at the original dataset:

```python
df = draw_map(df, 65, 40, -50, -140)
```

After normaizing the values, initialize the model and fit it to just the coordinates i.e. only location-based clustering:

```python
cluster_df = StandardScaler().fit_transform(np.nan_to_num(df[["xm", "ym"]]))

model = DBSCAN(eps=0.15, min_samples=10)
model.fit(cluster_df)
```

Get the predicted clusters and add them to the dataset:

```python
df["labels"] = model.labels_
```

Let's see how the clusters look:

```python
df = draw_map(df, 65, 40, -50, -140, clustered=True, n_clusters=len(set(model.labels_)), model=model)
```

This time, let's fit the model to both location and temperature data:

```python
cluster_df = StandardScaler().fit_transform(np.nan_to_num(df[["xm", "ym", "Tx", "Tm", "Tn"]]))

model = DBSCAN(eps=0.3, min_samples=10)
model.fit(cluster_df)
```

Add the predicted clusters to the dataset and visualize the clusters:

```python
df["labels"] = model.labels_
df = draw_map(df, 65, 40, -50, -140, clustered=True, n_clusters=len(set(model.labels_)), model=model)
```

Finally, save the model:

```python
pickle.dump(model, open("../saved_models/model_11.sav", "wb"))
```
