# Content-Based Recommender System

To recommend movies to a user based on their user profile

Import necessary modules:

```python
import pandas as pd
```

Get the dataset and see a sample (we'll start with movies dataset):

```python
movies_df = pd.read_csv("../datasets/movies.csv")
ratings_df = pd.read_csv("../datasets/ratings.csv")
movies_df.head()
```

Preprocess the dataset to put the movie's year of release in a separate column and convert the movie genres column entries into lists:

```python
movies_df["year"] = movies_df["title"].str.extract("(\(\d\d\d\d\))", expand=False)
movies_df["year"] = movies_df["year"].str.extract("(\d\d\d\d)", expand=False)
movies_df["title"] = movies_df["title"].str.replace("(\(\d\d\d\d\))", '')
movies_df["title"] = movies_df["title"].apply(lambda x: x.strip())
movies_df["genres"] = movies_df["genres"].str.split('|')
movies_df.head()
```

Create separate entries for every genre (this will be important for creating genre table):

```python
movies_with_genres_df = movies_df.copy()
for i, row in movies_df.iterrows():
    for genre in row["genres"]:
        movies_with_genres_df.at[i, genre] = 1
movies_with_genres_df = movies_with_genres_df.fillna(0)
movies_with_genres_df.head()
```

Now, let's look at the ratings dataset:

```python
ratings_df.head()
```

Process the dataset to remove the timestamp column:

```python
ratings_df = ratings_df.drop("timestamp", 1)
ratings_df.head()
```

Let's create a dummy user for whom we'll make recommendations:

```python
test_movies = pd.DataFrame([
    {'title':'Breakfast Club, The', 'rating':5},
    {'title':'Toy Story', 'rating':3.5},
    {'title':'Jumanji', 'rating':2},
    {'title':"Pulp Fiction", 'rating':5},
    {'title':'Akira', 'rating':4.5}
])
test_movies
```

Get the IDs of the movies the user has watched:

```python
test_ids = movies_df[movies_df["title"].isin(test_movies["title"].tolist())]
test_movies = pd.merge(test_ids, test_movies).drop("genres", 1).drop("year", 1)
test_movies
```

Use the above dataset to create a dataset of genres of the movies the user has watched:

```python
prefs_df = movies_with_genres_df[movies_with_genres_df["movieId"].isin(test_movies["movieId"].tolist())]
prefs_df = prefs_df.reset_index(drop=True).drop("movieId", 1).drop("title", 1).drop("genres", 1).drop("year", 1)
prefs_df
```

Now, create a user profile by taking a dot product of the user's watched genres dataset with the ratings dataset:

```python
user_profile = prefs_df.transpose().dot(test_movies["rating"])
user_profile
```

Create a genres table from the dataset we made earlier:

```python
genre_table = movies_with_genres_df.set_index(movies_with_genres_df["movieId"])
genre_table = genre_table.drop("movieId", 1).drop("title", 1).drop("genres", 1).drop("year", 1)
genre_table.head()
```

Take the weighted average of the genres table with the user profile to find out the preferences of the user:

```python
weighted_profile = ((genre_table*user_profile).sum(axis=1)/user_profile.sum()).sort_values(ascending=False)
weighted_profile.head()
```

The top 20 (or 10, the number may change) entries in the weighted average profile are the movies we should recommend to the user:

```python
recommendations_df = movies_df.loc[movies_df["movieId"].isin(weighted_profile.head(20).keys())]
recommendations_df
```

And that's it! We created a movie recommender engine without using any machine learning library :D
